```python
import pandas as pd 
df = pd.read_csv('data_500_days.csv')
df['Date'] = pd.to_datetime(df['Date']) 
# assume the following: 
# Cradle to gate emissions for PLA strands = 3.9 kg Co2eq / kg https://www.researchgate.net/publication/393510051_Sustainability_Assessment_of_the_3D-_Recycler_Process_Local_Plastic_Recycling_for_3D_Printing
# Cradle to gate emissions for TPU strands = 4.5 kg Co2eq/kg source: trust me bro 
# Cradle to gate emissions for PVA strands = 5.5 kg Co2eq/kg source: trust me bro
emissions_data_daily = {'Date':df['Date'],
                    'PLA_emissions':df['PLA_weight']*3.9*1e-3,
                    'PVA_emissions':df['PLA_weight']*5.5*1e-3,
                    'TPU_emissions':df['PLA_weight']*4.5*1e-3
                }
emissions_data_daily=pd.DataFrame(emissions_data_daily).set_index('Date')
# how do I get weekly - add up daily emissions. 
# dates=emissions_data_daily["Date"]
# emissions_data_daily=emissions_data_daily.drop(['Date'],axis=1)
# week_starts=dates.groupby(dates.index//7).min() # gets the week starts. 
# emissions_data_weekly=emissions_data_daily.groupby(emissions_data_daily.index // 7).sum()
# emissions_data_weekly['Week']=week_starts
# emissions_data_weekly = emissions_data_weekly.set_index('Week')
emissions_data_weekly = emissions_data_daily.resample('W-MON').sum()
emissions_data_monthly = emissions_data_daily.resample('MS').sum()
emissions_data_monthly

```
